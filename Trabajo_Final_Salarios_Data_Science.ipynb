Introduccion al Proyecto Data Set

Contexto Los " salarios de la ciencia de datosEl conjunto de datos proporciona información valiosa sobre las tendencias de compensación y las variaciones en el campo de la ciencia de datos de 2020 a 2023. Este conjunto de datos abarca una colección completa de información salarial de varias industrias, organizaciones y regiones geográficas, lo que permite a los profesionales de datos, investigadores y organizaciones analizar y comprender el panorama salarial predominante en el dominio de la ciencia de datos durante este período de cuatro años. El conjunto de datos sirve como un recurso valioso para las personas que buscan orientación profesional, las empresas que buscan comparar sus estrategias de compensación,e investigadores que investigan la dinámica en evolución del mercado laboral de ciencia de datos.

Contenido El conjunto de datos incluye información sobre el cargo, el tipo de empleo, el nivel de experiencia, el nivel de experiencia, el salario, la moneda del salario, la ubicación de la empresa, el salario en USD, la residencia del empleado, el tamaño de la empresa y el año. Esta información proporciona una visión completa de los detalles relacionados con el trabajo, los niveles de compensación, las características de la empresa y los aspectos temporales, lo que ofrece información valiosa para analizar y comprender el conjunto de datos.

Glosario:

Job Title: Define cuál es el rol de la persona.

Employment Type: Tipo de empleo Indica la naturaleza del empleo, como a tiempo completo, a tiempo parcial, por contrato o independiente.

Experience Level: Refleja la etapa de carrera de la persona o el nivel de experiencia profesional.

Expertise Leve: Nivel de experiencia Describe el nivel de experiencia de la persona.

Salary: Salario La cantidad que se paga por la persona en su moneda local.

Salary Currency: La moneda específica en la que se paga el salario de la persona.

Company Location: Residencia del empleado El país donde reside el empleado.

Salary in USD: Salario convertido a Dólares Estadounidenses (USD). Esta es también la variable objetivo.

Employee Residence: Residencia del empleado El país donde reside el empleado

Company size: Tamaño de la empresa Describe la magnitud de la empresa empleadora, clasificada en pequeña, mediana o grande.

```python
# Calcula la media de la columna
salary_mean = df['Salary'].mean()
salaryinUSD_mean = df['Salary in USD'].mean()
year_mean = df['Year'].mean()

# Calcula la moda de la columna
salary_mode = df['Salary'].mode().iloc[0]
salaryinUSD_mode = df['Salary in USD'].mode().iloc[0]
year_mode = df['Year'].mode().iloc[0]

# Imputa los valores nulos de la media
df['Salary'].fillna(salary_mean, inplace=True)
df['Salary in USD'].fillna(salaryinUSD_mean, inplace=True)
df['Year'].fillna(year_mean, inplace=True)

# Imputa los valores nulos de la moda
df['Salary'].fillna(salary_mode, inplace=True)
df['Salary in USD'].fillna(salaryinUSD_mode, inplace=True)
df['Year'].fillna(year_mode, inplace=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Seleccionar las columnas que serán las variables independientes
independent_variables = df[['Salary', 'Salary in USD', 'Year']]

# Codificar la columna 'Brand' utilizando one-hot encoding
independent_variables_encoded = pd.get_dummies(independent_variables, columns=['Year'])

# Seleccionar la columna que será la variable dependiente
dependent_variable = df['Salary']

# Dividir el conjunto de datos en train y test
X_train, X_test, y_train, y_test = train_test_split(independent_variables_encoded, dependent_variable, test_size=0.2, random_state=42)

# Crear el modelo de regresión lineal
model = LinearRegression()

# Ajustar el modelo con el conjunto de entrenamiento
model.fit(X_train, y_train)

# Hacer predicciones con el conjunto de prueba
predictions = model.predict(X_test)
```

### Evaluación del modelo

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)
print("Error Cuadrático Medio (MSE):", mse)
```

```python
import numpy as np


# Calcular la varianza natural de la columna 'Salary'
variance_price = np.var(df['Salary'])
print("Varianza natural de Salary:", variance_price)

# Visualizar el histograma de 'Salary'
plt.hist(df['Salary'], bins=30, edgecolor='k')
plt.xlabel('Salary')
plt.ylabel('Frecuencia')
plt.title('Histograma de Salary')
plt.show()
```

Dado que el MSE es una medida de la discrepancia entre las predicciones y los valores reales, un MSE más bajo indica que el modelo tiene un mejor ajuste y realiza predicciones más precisas

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, predictions)
print("Coeficiente de Determinación (R²):", r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV

# Definir el espacio de búsqueda de hiperparámetros
param_grid = {
    'fit_intercept': [True, False]
}

# Crear el modelo de regresión lineal
model = LinearRegression()

# Realizar la búsqueda de hiperparámetros con validación cruzada
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Obtener los mejores hiperparámetros y el mejor modelo
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluar el modelo final en el conjunto de prueba
predictions = best_model.predict(X_test)
```

```python
# Evaluar el modelo final en el conjunto de prueba
predictions = best_model.predict(X_test)

# Calcular el Error Cuadrático Medio (MSE) en el conjunto de prueba
mse = mean_squared_error(y_test, predictions)
print("Error Cuadrático Medio (MSE) del modelo final:", mse)

# Calcular el Coeficiente de Determinación (R²) en el conjunto de prueba
r2 = r2_score(y_test, predictions)
print("Coeficiente de Determinación (R²) del modelo final:", r2)
```
